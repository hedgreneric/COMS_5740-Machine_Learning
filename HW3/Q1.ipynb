```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

data = np.array([
    [0, 0], [10, 10], [100, 100], [1, 1], [0, 5], 
    [-3, 4], [50, -50], [-1, 1], [0, -10], [30, -4]
])

kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(data)

labels = kmeans.labels_
centroids = kmeans.cluster_centers_

plt.scatter(data[labels == 0, 0], data[labels == 0, 1], c='black', label='Cluster 1', marker='.')
plt.scatter(data[labels == 1, 0], data[labels == 1, 1], c='blue', label='Cluster 2', marker='.')

plt.scatter(centroids[:, 0], centroids[:, 1], s=150, c='red', marker='*', label='Centroids')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('KMeans Clustering (K=2)')
plt.legend()
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

mu = np.array([[(110/3), (110/3)], [(48/3), (-40/3)], [(29/3), (-13/3)]])

pts = np.array([[0, 0], [10, 10], [100, 100], [50, -50], [30, -4], 
                [1, 1], [0, 5], [-3, 4], [-1, 1], [0, 10]])


# Plot the points and centroids
plt.figure(figsize=(10, 6))

# Plot the points
plt.scatter(pts[:, 0], pts[:, 1], color='blue', label='Data Points')

# Plot the centroids
plt.scatter(mu[:, 0], mu[:, 1], color='red', marker='*', s=200, label='Centroids')

# Add labels and title
plt.title('K-means Clustering: Points and Centroids')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.axis('equal')

# Show the plot
plt.show()
```
