```python
import matplotlib.pyplot as plt
import numpy as np

X = np.array([[0, 2], [0.4, 1], [0.6, 1], [1, 0]])
y = np.array([-1, -1, 1, 1])

plt.scatter(X[y == -1, 0], X[y == -1, 1], color='red', label='Class -1')
plt.scatter(X[y == 1, 0], X[y == 1, 1], color='blue', label='Class 1')

plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.title('Scatter plot of data points')
plt.grid(True)
plt.show()
```

```python
w = np.array([1, -1])
b = 0.5

distances = np.abs(np.dot(X, w) + b) / np.linalg.norm(w)

min_distance = np.min(distances)
min_index = np.argmin(distances)

print(f"The closest sample is at index {min_index} with coordinates {X[min_index]} and the distance is {min_distance}")
```

```python
scaling_factor = 1 / (y[min_index] * (np.dot(X[min_index], w) + b))

w_scaled = w * scaling_factor
b_scaled = b * scaling_factor

inverse_norm_w = 1 / np.linalg.norm(w_scaled)

print(f"The scaled w is {w_scaled}, the scaled b is {b_scaled}")
print(f"1/||w||: {inverse_norm_w}, Minimum distance: {min_distance}")
```
